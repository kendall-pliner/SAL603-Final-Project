```python
#allow output from every line
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import numpy as np
import pandas as pd
```

```python
#Load all 4 CSV Files
Brewers = pd.read_csv("Brewers_2025_Roster.csv", index_col = 'Player')
Cubs = pd.read_csv("Cubs_2025_Roster.csv", index_col = 'Player')
Cardinals = pd.read_csv("Cardinals_2025_Roster.csv", index_col = 'Player')
Reds = pd.read_csv("Reds_2025_Roster.csv", index_col = 'Player')
```

```python
#Create 1 All-Inclusive Table
Combined = pd.concat([Brewers,Cubs,Cardinals,Reds])
```

```python
#Save Combined Table as CSV
Combined.to_csv("National_League_Central_4of5_Rosters.csv")
```
